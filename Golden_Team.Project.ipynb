```python
#Data explanation



# Invoice ID:
# Description: A unique identifier for each sale.
# Data Type: object (string), which is a text identifier for each sale.

# Branch:
# Description: The name of the branch where the sale took place.
# Data Type: object (string), indicating the location of the sale.

# City:
# Description: The city where the branch is located.
# Data Type: object (string), defining the geographical location of the branch.

# Customer type:
# Description: The type of customer, whether a regular or a loyal customer.
# Data Type: object (string), distinguishing the customer type to determine the type of service or promotion.

# Gender:
# Description: The gender of the customer.
# Data Type: object (string), useful for analyzing demographic data of customers.

# Product line:
# Description: The product category or type of the sold product.
# Data Type: object (string), identifying the types of products sold.

# Unit price:
# Description: The price per unit of the product.
# Data Type: float64, indicating the individual price of the product.

# Quantity:
# Description: The quantity of the product sold.
# Data Type: int64, representing the number of units sold.

# Tax 5%:
# Description: The value of the 5% tax on sales.
# Data Type: float64, representing the tax added to the total price.

# Total:
# Description: The total price of the invoice, including base price and taxes.
# Data Type: float64, the final amount of the invoice after taxes.

# Date:
# Description: The date of the sale.
# Data Type: object (string), which needs to be converted to a datetime type for better analysis.

# Time:
# Description: The time of the sale.
# Data Type: object (string), which needs to be converted to a time type for better analysis.

# Payment:
# Description: The payment method used in the transaction (e.g., cash, credit card).
# Data Type: object (string), specifying the payment method used by the customer.

# COGS (Cost of Goods Sold):
# Description: The cost of the goods sold, related to production or purchasing the product.
# Data Type: float64, representing the production or purchase cost of the product.

# Gross margin percentage:
# Description: The percentage of gross margin on sales.
# Data Type: float64, indicating the percentage of gross profit as a percentage of revenue.

# Gross income:
# Description: The gross income or profit after deducting the cost of goods sold.
# Data Type: float64, representing the gross profit achieved from the sale.

# Rating:
# Description: The customer's rating of the product or service.
# Data Type: float64, which could be a numerical rating used to gauge customer satisfaction.

# Summary:
# Text Columns (object): Include categorical information such as invoice ID, branch name, city, customer type, gender, product line, and payment method.
# Numeric Columns (float64, int64): Include quantitative and financial information such as unit price, quantity, tax, total price, cost of goods, gross margin, and gross income.
# Temporal Columns (datetime, time): Pertaining to the date and time of the sale.
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use("seaborn")
import seaborn as sns 
import datetime as dt
```

```python
data = pd.read_csv("supermarket_sales - Sheet1.csv")
df = data.copy()
```

```python
# تحويل عمود Date إلى datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')


print(df.dtypes)
```

```python
df['Total price'] = df['Unit price'] * df['Quantity'] 
df.head()
```

```python
plt.figure(figsize=(6,4))
mean = np.mean(df['Total price'])
print(mean)
plt.hist(df['Total price'])
plt.axvline(mean, color="r")
plt.show()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
df['Season'] = df['Date'].dt.month.apply(get_season)
```

```python
df.info()
```

```python
# حساب إجمالي المبيعات لكل موسم
sales_by_season = df.groupby('Season')['Total'].sum()

# عرض النتائج
print("Total sales by season:")
print(sales_by_season)
```

```python
# Sales Analysis by Demand 
# Objective: To know the impact of each payment method on sales volume. This analysis can help to understand the preferences of customers well and to confirm payment


# Calculate total sales per payment method
sales_by_payment_method = df.groupby('Payment')['Total'].sum()

# Show results
print("Total sales by payment method:")
print(sales_by_payment_method)
```

```python
# Profit Analysis
# Objective: Measure gross profit and net profit to understand how costs and taxes impact profits. This analysis can help evaluate financial performance across different branches of a company or product categories.

# Analysis Steps:
# Calculating Gross Profit: Gross profit is the difference between total sales and costs of goods sold (cogs).
# Calculating Net Profit: This is the total profit after taxes are deducted.



df['Profit'] = df['Total'] - df['cogs']

# Calculate the total profit for each row
profit_by_branch = df.groupby('Branch')['Profit'].sum()


# Calculate total profit for each product category
profit_by_product_line = df.groupby('Product line')['Profit'].sum()

print("Total profit by branch:")
print(profit_by_branch)

print("\nTotal profit by product line:")
print(profit_by_product_line)
```

```python
# Calculate total sales for each customer type.
sales_by_customer_type = df.groupby('Customer type')['Total'].sum()

# Calculate average sales per customer type
average_sales_by_customer_type = df.groupby('Customer type')['Total'].mean()

# Show results
print("Total sales by customer type:")
print(sales_by_customer_type)

print("\nAverage sales by customer type:")
print(average_sales_by_customer_type)
```

```python
# Sales Analysis by Season
# Objective: Understand how seasons affect sales, and identify the seasons that have the highest and lowest sales. This analysis can help improve sales and stocking strategies according to the seasons.

# Analysis Steps:
# Calculate Total Sales for Each Season: Determine the sales volume for each season.

# Calculate Sales Percentage: Analyze the percentage of sales for each season from total sales.

# Implement Analysis

# Calculate Total Sales for Each Season
sales_by_season = df.groupby('Season')['Total'].sum()

# Total Sales Calculation
total_sales = df['Total'].sum()

# Calculate sales percentage for each season
sales_percentage_by_season = (sales_by_season / total_sales) * 100

# Show results
print("Total sales by season:")
print(sales_by_season)

print("\nSales percentage by season:")
print(sales_percentage_by_season)

# Show total sales by season
sales_by_season.plot(kind='bar', color='lightgreen')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.title('Total Sales by Season')
plt.xticks(rotation=45)
plt.show()

# Show sales percentage by season
sales_percentage_by_season.plot(kind='bar', color='lightcoral')
plt.xlabel('Season')
plt.ylabel('Sales Percentage')
plt.title('Sales Percentage by Season')
plt.xticks(rotation=45)
plt.show()
```

```python

# Calculate total sales per gender
sales_by_gender = df.groupby('Gender')['Total'].sum()

# Total Sales Calculation
total_sales = df['Total'].sum()

# Calculate sales percentage for each gender
sales_percentage_by_gender = (sales_by_gender / total_sales) * 100


print("Total sales by gender:")
print(sales_by_gender)

print("\nSales percentage by gender:")
print(sales_percentage_by_gender)

# Show total sales by gender
sales_by_gender.plot(kind='bar', color='lightblue')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.title('Total Sales by Gender')
plt.xticks(rotation=0)
plt.show()


# Show sales percentage by gender
sales_percentage_by_gender.plot(kind='bar', color='lightcoral')
plt.xlabel('Gender')
plt.ylabel('Sales Percentage')
plt.title('Sales Percentage by Gender')
plt.xticks(rotation=0)
plt.show()
```

```python


# Convert Date column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.to_period('M')

# Calculate total sales for each month and year
monthly_sales = df.groupby('Month')['Total'].sum().reset_index()

# Convert month period to text string for easier drawing
monthly_sales['Month'] = monthly_sales['Month'].astype(str)


plt.figure(figsize=(12, 6))


plt.plot(monthly_sales['Month'], monthly_sales['Total'], marker='o', linestyle='-', color='b')


plt.xticks(rotation=90) 
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python


# Convert date column to datetime type
df['Date'] = pd.to_datetime(df['Date']) 

# Filter data for February only
february_sales = df[df['Date'].dt.month == 2]

# Calculate total sales for February
total_february_sales = february_sales['Total'].sum()
print(f"Total Sales in February: {total_february_sales}")

# Calculate total sales for each product category in February.
sales_by_product_line_feb = february_sales.groupby('Product line')['Total'].sum()
print("\nSales by Product Line in February:")
print(sales_by_product_line_feb)

# Calculate total sales for each month
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total'].sum().reset_index()
monthly_sales['Date'] = monthly_sales['Date'].astype(str)  # تحويل الفترة الزمنية إلى سلسلة نصية
print("\nMonthly Sales:")
print(monthly_sales)
```

```python
# Identify the best and lowest selling times: Analyze time data to identify peak and low sales hours.


# Add today's hour column
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))
# حساب مبيعات حسب ساعة اليوم
hourly_sales = df.groupby('Hour')['Total'].sum()

# Sales Chart by Hour of the Day
hourly_sales.plot(kind='line', marker='o')
plt.title('Sales by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

```

```python
# Drawing ratings against sales
sns.scatterplot(data=df, x='Rating', y='Total')
plt.title('Sales vs Customer Rating')
plt.xlabel('Customer Rating')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
